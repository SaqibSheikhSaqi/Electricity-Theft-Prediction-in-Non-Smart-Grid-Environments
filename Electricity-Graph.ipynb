```python
import pandas as pd

TheftData=pd.read_csv("D:\ElectricityTheft\dataset\TunedElectricityDataset.csv")
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
X = TheftData.iloc[:, 3:1036].astype('float32' )
y = TheftData.iloc[:, 1]
```

```python
import pandas as pd
import matplotlib.pyplot as plt

title="Mean Electricity Consumption for Theft and Not-Theft in 2014"  
xlbl="Year-2014"  
date_columns = [col for col in TheftData.columns if "2014" in col]

# Calculate mean for FLAG = 0 and FLAG = 1
mean_0 = TheftData[TheftData["FLAG"] == 0][date_columns].mean().mean()  # Mean for FLAG 0
mean_1 = TheftData[TheftData["FLAG"] == 1][date_columns].mean().mean()  # Mean for FLAG 1

# Reference value for calculation of percentage
reference_value = 17.5  # You can choose any reference value

# Calculate percentages
percentage_0 = (mean_0 / reference_value) * 100
percentage_1 = (mean_1 / reference_value) * 100
plt.figure(figsize=(4, 4))
# Plotting
plt.bar(["Not-Theft (0)", "Theft (1)"], [percentage_1, percentage_0, ], width=0.1, align='edge')

# Adding labels and title
plt.xlabel(xlbl)
plt.ylabel("Mean Electricity Consumption (%)")
#plt.title(title)

# Show plot
plt.tight_layout()
plt.show()
```
